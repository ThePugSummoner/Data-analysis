```python
import pandas as pd
import pandas_gbq

projectid = "decisive-circle-378616"

sql = """
SELECT * FROM `decisive-circle-378616.test_data2023.G4_test_20230222`; 
"""

df_comm = pd.read_gbq(sql,projectid)

df_comm.head()
```

```python
df_comm.groupby(by='location').count().head(50)

                                
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
plt.figure(figsize=(12,6), dpi=150)

plt.plot(df_comm.index, df_comm['total_events'])

plt.grid(True)

plt.title('Number of events per user')

plt.xlabel('User')
plt.ylabel('Number of events')

plt.xlim(-100,24000)
plt.ylim(0, 1000)

plt.yticks(np.arange(0, 1000, 100))

plt.xticks(np.arange(0, 24000, 1500))

plt.show()
```

```python
plt.figure(figsize=(10,5), dpi=100)

plt.hist(df_comm['revenue'],  bins=np.arange(0,350,10), color='orange')

plt.title('Distribution of revenue')
plt.xlabel('Revenue')
plt.ylabel('Number of users')
plt.yticks(np.arange(0, 25, 2))
plt.xticks(np.arange(0, 350, 20))
plt.grid(True)
plt.xlim(0,350)

plt.show()
```

```python
plt.figure(figsize=(10,5), dpi=100)

plt.hist(df_comm['location'])
plt.title('Distributions of countries')
plt.xlabel('Country')
plt.ylabel('Number of users')

plt.xticks(np.arange(0, 110,5),rotation=90)
plt.xlim(-1,110)
plt.grid(True)

plt.show()
```
