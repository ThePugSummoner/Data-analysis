## Visualization


```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Read gapminder data from plotly

df = px.data.gapminder()
```

```python
df
```

```python
df.info()
```

```python
# Line plot of GDP per capita for each year in Finland and Sweden

df_FI = df[df['country'] == 'Finland']

df_FI

df_SWE = df[df['country'] == 'Sweden']

df_SWE
```

```python
plt.plot(df_FI['year'], df_FI['gdpPercap']) #plt.plot(x,y)
plt.plot(df_SWE['year'], df_SWE['gdpPercap']) #plt.plot(x,y)
```

```python
plt.plot(df_SWE['year'], df_SWE['gdpPercap']) #plt.plot(x,y)
```

```python
# Create a simple plot visualization (x-axis and y-axis)

plt.figure(figsize=(12,4), dpi=150) #picture size 

plt.plot(df_FI['year'], df_FI['gdpPercap'], marker='.', markersize=12) #plt.plot(x,y)
plt.plot(df_SWE['year'], df_SWE['gdpPercap'], marker='.', markersize=12) #plt.plot(x,y)

#plt.scatter(df_FI['year'], df_FI['gdpPercap']) #plt.plot(x,y)
#plt.scatter(df_SWE['year'], df_SWE['gdpPercap']) #plt.plot(x,y)

plt.title('GDP per capita in Finland and Sweden') # Header

plt.xlabel('Years') # X-axis title
plt.ylabel('GDP per capita') # Y-axis title

plt.legend(['FI', 'SWE']) # Legend of plot

plt.grid(True) # Grid#

plt.xlim(1950,2010) # X-axis grid adjustment
plt.ylim(5000, 35000) # Y-axis grid adjustment

#plt.xticks(df_FI['year'])
plt.xticks(np.arange(1950, 2010, 5)) # x-axis ticks came with 5 years interval

plt.show() # End of demonstration
```

```python
# Visualize three parameters using scatter plot

# Create a simple plot visualization (x-axis and y-axis)

plt.figure(figsize=(6,6), dpi=150)

plt.scatter(df_FI['year'], df_FI['gdpPercap'], c = df_FI['lifeExp'], cmap='RdBu') # Scatter plot

plt.title('GDP per capita in Finland') # Header

plt.xlabel('Years') # X-axis title
plt.ylabel('GDP per capita') # Y-axis title

plt.colorbar()

plt.grid(True) # Grid#

plt.show() # End of demonstration
```

```python
# Bar chart

plt.figure(figsize=(12,4), dpi=150) # Plot size 

plt.bar(df_FI['year'] - 0.5, df_FI['gdpPercap'], width=1)
plt.bar(df_SWE['year'] + 0.5, df_SWE['gdpPercap'], width=1)

#plt.scatter(df_FI['year'], df_FI['gdpPercap']) #plt.plot(x,y)
#plt.scatter(df_SWE['year'], df_SWE['gdpPercap']) #plt.plot(x,y)

plt.title('GDP per capita in Finland and Sweden') # Header

plt.xlabel('Years') # X-axis title
plt.ylabel('GDP per capita') # Y-axis title

plt.legend(['FI', 'SWE']) # Legend of plot

plt.grid(True) # Grid#

plt.xlim(1950,2010) # X-axis grid adjustment
plt.ylim(0, 35000) # Y-axis grid adjustment

#plt.xticks(df_FI['year'])
plt.xticks(np.arange(1952, 2012, 5)) # x-axis ticks came with 5 years interval

plt.show() # End of demonstration
```

```python
# Stacked Bar chart

plt.figure(figsize=(12,4), dpi=150) # Plot size 

plt.bar(df_FI['year'], df_FI['gdpPercap'], width=1) # Bars adjustment -0.5
plt.bar(df_SWE['year'], df_SWE['gdpPercap'], bottom=df_FI['gdpPercap'],width=1) # Bars adjustment +0.5

plt.title('GDP per capita in Finland and Sweden') # Header

plt.xlabel('Years') # X-axis title
plt.ylabel('GDP per capita') # Y-axis title

plt.legend(['FI', 'SWE']) # Legend of plot

plt.grid(True) # Grid#

plt.xlim(1950,2010) # X-axis grid adjustment
plt.ylim(0, 70000) # Y-axis grid adjustment

#plt.xticks(df_FI['year'])
plt.xticks(np.arange(1952, 2012, 5)) # x-axis ticks came with 5 years interval

plt.show() # End of demonstration
```
