```python
import pandas as pd
```

```python
df = pd.read_csv('cars_for_exercise.csv')

df.info()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

for i in df.columns: #Goes through all columns in the dataframe

    if df[i].dtype.kind in 'iuf': #Accept numerical data

        print('Data column ', i) #Print column header, VehicleClass, RegData, etc.
 
        print('The maximum of ',i,' is', df[i].max()) #Print column maximum

        print('The minimum of ',i,' is', df[i].min()) #Print column minimum

        print('The average of ',i,' is', df[i].mean()) #Print column average 

        print('The standard deviation of ',i,' is', df[i].std()) #Print column stadard deviation

        print('The number of missing data points in ',i,' is', df[i].isna().sum()) #Print number of missing values

        print('The percent of missing data in ',i,' is', round((df[i].isna().sum())/len(df)*100,2)) #Precent of missing data

        print('The value count of ',i,' is:') 

        print(df[i].value_counts()) #Print the occurence rate of each value

        #Figures

        plt.figure(figsize=(10,6), dpi = 100) #Simple line plot of column

        plt.subplot(2,1,1)

        plt.plot(df[i])

        plt.title(['Values of ',i])

        plt.grid()

        plt.subplot(2,1,2) #Histogram of column values

        plt.hist(df[i])

        plt.title(['Histogram of ',i])

        plt.grid()

        plt.show()

    else:

        print('Data column ', i)

        print('The number of missing data points in ',i,' is', df[i].isna().sum())

        print('The percent of missing data in ',i,' is', round((df[i].isna().sum())/len(df)*100,2))

        vc = df[i].value_counts()

        print('There are',len(vc[vc>1]), ' values of ', i ,'which appear multiple times')

        print('')
    
        print('#######################################')
```
